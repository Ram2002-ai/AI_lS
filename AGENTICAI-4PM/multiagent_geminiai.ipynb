```python
import os
from crewai import Agent,Task,Crew
from crewai_tools import SerperDevTool

os.environ['GEMINI_API_KEY']=
os.environ['SERPER_API_KEY']=
```

```python
search_tool = SerperDevTool()

# -----------------------
# 👥 Agents
# -----------------------

teacher = Agent(
    role="Teacher",
    goal="Explain concepts clearly step by step",
    backstory="Experienced teacher who explains with examples",
    llm="gemini-2.5-flash",
    verbose=True
)

researcher = Agent(
    role="Researcher",
    goal="Find latest information",
    backstory="Expert in searching real-world data",
    llm="gemini-2.5-flash",
    tools=[search_tool],  # ✅ tool added
    verbose=True
)

simplifier = Agent(
    role="Simplifier",
    goal="Make things easy",
    backstory="Breaks complex ideas into simple language",
    llm="gemini-2.5-flash",
    verbose=True
)

student = Agent(
    role="Student",
    goal="Take notes",
    backstory="Writes simple notes like a learner",
    llm="gemini-2.5-flash",
    verbose=True
)

examiner = Agent(
    role="Examiner",
    goal="Create questions",
    backstory="Tests understanding",
    llm="gemini-2.5-flash",
    verbose=True
)

# -----------------------
# 📘 Topic
# -----------------------
topic = "What is Artificial Intelligence?"

# -----------------------
# 🧩 Tasks
# -----------------------

task1 = Task(
    description=f"Search and find 3 important points about {topic}",
    expected_output="3 clear points",
    agent=researcher
)

task2 = Task(
    description=f"Explain {topic} in simple steps with examples",
    expected_output="Step-by-step explanation",
    agent=teacher
)

task3 = Task(
    description=f"Simplify the explanation of {topic}",
    expected_output="Very easy explanation",
    agent=simplifier
)

task4 = Task(
    description=f"Write short notes on {topic} like a student",
    expected_output="Simple notes",
    agent=student
)

task5 = Task(
    description=f"Create 3 simple questions on {topic}",
    expected_output="3 easy questions",
    agent=examiner
)

# -----------------------
# 🤝 Crew
# -----------------------

crew = Crew(
    agents=[teacher, researcher, simplifier, student, examiner],
    tasks=[task1, task2, task3, task4, task5],
    verbose=True
)

# -----------------------
# 🚀 Run
# -----------------------

import asyncio

result = await crew.kickoff_async()
print(result)

#print("\n" + "="*50)
#print("FINAL OUTPUT")
#print("="*50 + "\n")
print(result)
```
