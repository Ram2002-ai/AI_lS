```python
#After restart session make sure import cell is executed again to run below code
loader=UnstructuredFileIOLoader("Python Oops Basic Functions.pdf")
```

```python
text_splitter = CharacterTextSplitter(chunk_size=2000, chunk_overlap=400)
```

```python
with open("Python Oops Basic Functions.pdf", "rb") as f:
  corrected_loader=UnstructuredFileIOLoader(f)
  document = corrected_loader.load()
texts=text_splitter.split_documents(document)
```

```python
embeddings=HuggingFaceEmbeddings()
```

```python
persist_dir='vector_db'
```

```python
llm=ChatGroq(model='llama-3.3-70b-versatile',temperature=2)
```

```python
#filtered_texts = filter_complex_metadata(texts)
vectordb=Chroma.from_documents(documents=texts, embedding=embeddings, persist_directory=persist_dir)
```

```python
retriver=vectordb.as_retriever()
```

```python
llm=ChatGroq(model='llama-3.3-70b-versatile',temperature=0)
```

```python
qa_chain=RetrievalQA.from_chain_type(llm=llm,chain_type="stuff",retriever=retriver, return_source_documents=True)
```

```python
query='classes'

response=qa_chain.invoke({'query':query})
```

```python
import gradio as gr
def chatbot_response(question):
  try:
    response=qa_chain.invoke({'query':question})
    answer=response["result"]
    sources=[]
    for doc in response["source_documents"]:
      if "source" in doc.metadata:
        sources.append(doc.metadata["source"])
    source_text="\n".join(list(set(sources)))
    final_response=f"""
{answer}
Sources:
{source_text}
"""
    return final_response
  except Exception as e:
    return f"Error: {str(e)}"


#Gradio UI
demo=gr.Interface(
    fn=chatbot_response,
    inputs=gr.Textbox(
        lines=2,
        placeholder="Ask a question from pdf...."
    ),
    #outputs="text",
    outputs=gr.Textbox(lines=10, label="Response"),
    title="RAG Chatbot to read from PDF file",
    description="LLMOPS project using Langchain + Groq + ChromaDB + Gradio",
    theme="Seafoam"
)
demo.launch(debug=True)
```
